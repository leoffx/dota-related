```
treasures = ['https://steamcommunity.com/market/priceoverview/?appid=570&currency=7&market_hash_name=Immortal%20Treasure%20I%202018','https://steamcommunity.com/market/priceoverview/?appid=570&currency=7&market_hash_name=Immortal%20Treasure%20II%202018',
             'https://steamcommunity.com/market/priceoverview/?appid=570&currency=7&market_hash_name=Immortal%20Treasure%20III%202018', 'https://steamcommunity.com/market/priceoverview/?appid=570&currency=7&market_hash_name=Trust%20of%20the%20Benefactor%202018']

price = [[] for _ in range(len(treasures))]
```

```
"""price[0] = np.load('1.npy')
price[1] = np.load('2.npy')
price[2] = np.load('3.npy')
price[3] = np.load('4.npy')"""
```

```
while True:
  clear_output()  
  fig = plt.figure(figsize=(12,12))
  try:
    for i in range(len(treasures)):
      r = requests.get(treasures[i])
      price[i].append(float(price_str(r.json()['lowest_price'])))
      fig.add_subplot(2 ,2 , i+1)
      plt.plot(price[i], 'r-')
      plt.title('Tesouro ' + str(i))
  except:
    pass

  plt.show()


  np.save('1', price[0])
  np.save('2',price[1])
  np.save('3', price[2])
  np.save('4', price[3])


  time.sleep(20) 
```
